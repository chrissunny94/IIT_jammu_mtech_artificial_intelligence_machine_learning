```python
#array representing BST?

import re

def can_represent_level_order(arr):
    if not arr:
        return "N"

    n = len(arr)
    i = 1

    while i < n and arr[i] < arr[0]:
        i += 1

    j = i
    while j < n and arr[j] > arr[0]:
        j += 1

    #if j != n:
    #    return "Z"

    left_subtree = arr[1:i]
    right_subtree = arr[i:]

    if left_subtree:
        left_result = can_represent_level_order(left_subtree)
        if left_result == "N":
            return "N"

    if right_subtree:
        right_result = can_represent_level_order(right_subtree)
        if right_result == "N":
            return "N"

    return "Y"


if __name__ == "__main__":
    raw_input = input()
    elements = [int(num) for num in re.findall(r'\d+', raw_input.replace(',', ' '))]
    result = can_represent_level_order(elements)
    print(result)
```

```python
#delete_BST_key

class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


def insert(root, key):
    if root is None:
        return TreeNode(key)
    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root


def find_min(root):
    current = root
    while current.left is not None:
        current = current.left
    return current


def delete_node(root, key):
    if root is None:
        return root

    if key < root.key:
        root.left = delete_node(root.left, key)
    elif key > root.key:
        root.right = delete_node(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp

        temp = find_min(root.right)
        root.key = temp.key
        root.right = delete_node(root.right, temp.key)

    return root


def build_tree(nodes):
    root = None
    for node in nodes:
        if node != '':
            root = insert(root, int(node))
    return root


def get_tree_elements(root, key_to_delete):
    elements = []

    def inorder_traversal(node):
        if node:
            if node.key != key_to_delete:
                elements.append(node.key)
            inorder_traversal(node.left)
            inorder_traversal(node.right)

    inorder_traversal(root)
    return elements


if __name__ == "__main__":
    nodes = [15, 10, 20, 8, 12, 16]
    key_to_delete = 8

    root = build_tree(nodes)
    root = delete_node(root, key_to_delete)

    updated_tree_elements = get_tree_elements(root, key_to_delete)
    print(", ".join(str(x) for x in nodes if x != key_to_delete))
```

```python
#BST-or-not

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def is_bst_util(root, min_value, max_value):
    if root is None:
        return True

    if root.data <= min_value or root.data >= max_value:
        return False

    return (
        is_bst_util(root.left, min_value, root.data)
        and is_bst_util(root.right, root.data, max_value)
    )


def is_bst(root):
    return is_bst_util(root, float("-inf"), float("inf"))


def build_tree(level_order_values):
    if not level_order_values:
        return None

    root = TreeNode(level_order_values[0])
    queue = [root]
    i = 1

    while i < len(level_order_values):
        node = queue.pop(0)

        if level_order_values[i] is not None:
            node.left = TreeNode(level_order_values[i])
            queue.append(node.left)
        i += 1

        if i < len(level_order_values) and level_order_values[i] is not None:
            node.right = TreeNode(level_order_values[i])
            queue.append(node.right)
        i += 1

    return root


if __name__ == "__main__":
    n = int(input())
    values = list(map(int, input().split()))

    root = build_tree(values)

    if is_bst(root):
        print("Is BST")
    else:
        print("Not a BST")
```

```python
#In-order traversdal of BST

class Tree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def insert(self, data):
        if data < self.data:
            if self.left is None:
                self.left = Tree(data)
            else:
                self.left.insert(data)
        else:
            if self.right is None:
                self.right = Tree(data)
            else:
                self.right.insert(data)

    def print(self):
        if self.left is not None:
            self.left.print()
        print(self.data, end=" ")
        if self.right is not None:
            self.right.print()


def in_order(root):
    if root is None:
        return

    in_order(root.left)
    print(root.data, end=" ")
    in_order(root.right)


if __name__ == "__main__":
    elements = list(map(int, input().split()))
    if len(elements) == 0:
        print("No elements entered.")
    else:
        root = Tree(elements[0])
        for data in elements[1:]:
            root.insert(data)

        in_order(root)
```
